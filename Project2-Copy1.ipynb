### Data Cleaning (Replace this with more specific notes!)

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

# Checking for duplicated data
sum(df.duplicated())
```

```python
# Dropping duplicated values

df.drop_duplicates(inplace=True)
```

```python
# Deleting the columns I am not going to use

to_delete = ['id', 'imdb_id', 'popularity', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 'overview', 'production_companies', 'vote_count', 'vote_average']


df.drop(columns=to_delete,inplace=True)

#Checking if columns were deleted properly
df.columns
```

```python
# Deleting movies with budget and revenue zero
# Calculating how many films
df.shape
```

```python
#Replacing zero with null
df["budget"] = df.budget.replace(0,np.NaN)
df["revenue"] = df.revenue.replace(0,np.NaN)
df["runtime"] = df.runtime.replace(0,np.NaN)

# Deleting NA
df.dropna(subset=["revenue","budget","runtime"],inplace=True)

#Checking how many films are left
df.shape
```

```python
# Spliting Genres and keeping only the first one

new = df.genres.str.split("|",n=1, expand=True)
df["genre"] = new[0]
df.drop(columns=["genres"], inplace=True)
df.head()
```

```python
#Checking if runtime looks ok
df.head().sort_values(by="runtime")
```

```python
#Checking data types
df.dtypes
```

```python
# Converting release date to date
df["release_date"] = pd.to_datetime(df.release_date,)

#Converting budget and revenue from float to int
df['budget'] = df['budget'].astype(int)
df['revenue'] = df['revenue'].astype(int)
```

```python
# Checking if conversion worked
df.dtypes
```

### 1. What are the most popular types of film genres?

```python
# Finding the count of all films per genres. The most common genre is Drama with 890 films, followed by Action 689

df1 = pd.crosstab(df.genre,df.revenue.count(),colnames=["#number_of_films"])
df1
```

### 2. Is Drama's popularity increasing or decreasing in a trendline?

```python
df_drama = df.query("genre == 'Drama'")
pd.crosstab(df_drama.release_year,df.revenue.count(),colnames=["#number_of_films"]).plot.line();


```

```python
# With the trendline chart above, we know there have been more drama films, but as a percentage of all the films?

#Normalizing the char
pd.crosstab(df_drama.release_year,df.revenue.count(),colnames=["#number_of_films"],normalize=True).plot.line();
```
